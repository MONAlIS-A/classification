## Dataset Overview

The dataset consists of 13 variables, each representing clinical features and demographic information of heart failure patients:

1. **age**: Age of the patient (years)
2. **anaemia**: Whether the patient has anaemia (0: no, 1: yes)
3. **creatinine_phosphokinase**: Level of the CPK enzyme in the blood (mcg/L)
4. **diabetes**: Whether the patient has diabetes (0: no, 1: yes)
5. **ejection_fraction**: Percentage of blood leaving the heart with each contraction (%)
6. **high_blood_pressure**: Whether the patient has high blood pressure (0: no, 1: yes)
7. **platelets**: Platelets in the blood (kiloplatelets/mL)
8. **serum_creatinine**: Level of serum creatinine in the blood (mg/dL)
9. **serum_sodium**: Level of serum sodium in the blood (mEq/L)
10. **sex**: Gender of the patient (1: male, 0: female)
11. **smoking**: Whether the patient smokes (0: no, 1: yes)
12. **time**: Follow-up period (days)
13. **DEATH_EVENT**: Whether the patient died during the follow-up period (0: no, 1: yes) [Target Variable]

```python
# Let's see how many patients died due to heart failures in the our dataset
patients_died = df.death_event.sum()
patients_survived = df.death_event.count() - df.death_event.sum()
print(f'{patients_died} Patients died due to heart failure')
print(f'{patients_survived} Patients survived after heart failure')
```

# Observation:
Out of 299 patients, 96 patients have died and 203 patients survived the heart failures.

```python
# Let us plot te result of patients who have died and those who have survived with distribution of their ages
df['patient_status'] =df.death_event.map({0: 'Survived', 1: 'Died'})
df['gender'] = df.sex.map({0: 'Female', 1: 'Male'})
pivot_table = df.pivot_table(index='patient_status', columns='gender', aggfunc='size', fill_value=0)
pivot_table.plot(kind='bar', stacked=True, title='Gender Wise Distribution of Patients Died Vs Patients Survived', xlabel='Patient Status', ylabel='Number of Patients', color = ['black', 'green'], alpha=0.3)
```

```python
# Let us group the ages in order to find out at what age patients have had their heart failures
bins = [0,40,50,60,70,80,90,100]
labels = ['below_40','41-50', '51-60', '61-70', '71-80', '81-90','91-100']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
age_group = df['age_group'].value_counts().reindex(labels)

# Let us plot a pie chart with different shades of blue color

colors = ['#00509E', 'green', 'red', 'yellow', 'violet', 'orange', '#CCE5FF']

plt.figure(figsize=(8,6))
plt.pie(age_group, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Patient with heart failure age groups ')
plt.legend(age_group.index, title='Age Groups', bbox_to_anchor=(1.05,1), loc='upper left')
plt.show()
```

# Observation: 
The above chart suggest that petients aged between 41 years of age to 80 years of age have had their heart failures.

```python
# Let's see which clinical feature is causing more heart failures
major_causes =df.groupby('patient_status')[['anaemia', 'creatinine_phosphokinase', 
                                                           'diabetes','ejection_fraction', 
                                                           'high_blood_pressure','serum_creatinine', 
                                                            'serum_sodium','smoking',
                                                           'platelets']].mean()
features = ['anaemia', 'creatinine_phosphokinase', 'diabetes',
           'ejection_fraction', 'high_blood_pressure','serum_creatinine', 
           'serum_sodium','smoking','platelets']

fig, axes = plt.subplots(2,5, figsize = (16,12))
axes = axes.flatten()

for i, feature in enumerate(features):
    axes[i].bar(major_causes.index, major_causes[feature], color='red', alpha=0.2)
    axes[i].set_title(feature)
    axes[i].set_xlabel('Death Event')
    axes[i].set_ylabel('Mean Value')

# remove empty subplots
for j in range(len(features), len(axes)):
    fig.delaxes(axes[j])
    
plt.tight_layout()
plt.show()
```

```python
# Defining our target variable
y = df.death_event
# Defining our feature variable X
features = ['anaemia', 'creatinine_phosphokinase', 'diabetes',
           'ejection_fraction', 'high_blood_pressure','serum_creatinine', 
           'serum_sodium','smoking','platelets']
X = df[features]
# Defining model
heart_failure_model = DecisionTreeRegressor(random_state=1)
heart_failure_model.fit(X,y)
```

```python
# Making prediction on 5 patients
print('Making prediction for first 5 patients')
print(X.head())
print('The predictions are')
print(heart_failure_model.predict(X.head()))
```

```python
# Mean Absolute Error
mean_absolute_error(y, heart_failure_model.predict(X))
```

```python
# Split the dataset to find out how the model performs on actual data against trained data
train_X, val_X, train_y, val_y = train_test_split(X,y, random_state=0)
heart_failure_model = DecisionTreeRegressor()
heart_failure_model.fit(train_X, train_y)
```

```python
# Let's print our model's predictions
print(mean_absolute_error(val_y, heart_failure_model.predict(val_X)))
```

Its levels in the blood can provide important diagnostic information. It is proved from the above visualization that higher level of creatinine_phosphokinase kills more people. -- diabetes: is a chronic condition where the body either does not produce enough insulin (Type 1 diabetes) or does not use insulin effectively (Type 2 diabetes). Managing diabetes is crucial as it can have significant effects on overall health, including cardiovascular health. The people with diabetes are more prone to heart failures. The diabetic people have equal chance of surviving heart failure or loosing their lives. -- ejection_fraction (EF): is a key metric used in cardiology to assess how well the heart is pumping blood. Specifically, it measures the percentage of blood that is ejected from the left ventricle of the heart with each contraction. The left ventricle is the heart's main pumping chamber, and the ejection fraction provides insight into the heart's overall function. In the heart failure dataset, ejection_fraction is likely one of the features used to predict patient outcomes. It can be an important predictor, as patients with significantly reduced ejection fractions are at a higher risk of adverse outcomes. -- high_blood_pressure: High Blood Pressure (Hypertension) is a common condition where the force of the blood against the walls of your arteries is consistently too high. Over time, this can lead to serious health problems, such as heart disease, stroke, and kidney failure. Patients with hypertension may have an increased risk of adverse outcomes due to the added strain on their cardiovascular system. -- serum_creatinine: is a blood test that measures the amount of creatinine, a waste product produced by muscle metabolism, in the bloodstream. The kidneys filter creatinine out of the blood, and it's excreted in urine. Higher serum creatinine levels in patients may correlate with worse outcomes, as they may indicate reduced kidney function, which is often associated with more advanced heart failure. -- serum_sodium: Serum Sodium refers to the concentration of sodium in the blood. Sodium is an essential electrolyte that helps regulate water balance in the body, nerve function, and muscle contraction. Analyzing the relationship between serum sodium levels and patient outcomes might help in understanding the role of electrolyte balance in the progression of heart failure. -- smoking: is a significant risk factor for many cardiovascular diseases, including heart failure. Smoking introduces harmful chemicals into the body that can damage the heart and blood vessels, leading to a range of health issues. Analyzing the relationship between smoking status and heart failure progression can provide insights into the role of lifestyle factors in managing heart failure. -- platelets: are a crucial component of blood that helps with clotting. In the context of heart failure, platelet levels can provide important insights into the patient's condition and overall cardiovascular health. Analyzing how platelet levels correlate with patient status (e.g., survived or died) and other clinical features can help in understanding its role in heart failure.
